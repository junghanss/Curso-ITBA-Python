```python
import pandas as pd, numpy as np
from datetime import datetime,timedelta
from matplotlib import pyplot as plt      # Del modulo pyplot usaremos lo necesario para los graficos
from matplotlib import dates as mpl_dates # Del modulo dates usaremos el formato especifico para la clase date/time
import requests # Para definir nuestro propio wget
```

```python
def wget(url):
    r = requests.get(url, allow_redirects=True)
    with open(url[url.rfind('/') + 1::], 'wb') as f:
        f.write(r.content)

wget("https://raw.githubusercontent.com/junghanss/curso_itba/master/EDN.csv")
wget("https://raw.githubusercontent.com/junghanss/curso_itba/master/YPF.csv")
wget("https://raw.githubusercontent.com/junghanss/curso_itba/master/BBAR.csv")
wget("https://raw.githubusercontent.com/junghanss/curso_itba/master/BMA.csv")
wget("https://raw.githubusercontent.com/junghanss/curso_itba/master/GGAL.csv")
wget("https://raw.githubusercontent.com/junghanss/curso_itba/master/IRS.csv")
wget("https://raw.githubusercontent.com/junghanss/curso_itba/master/CRESY.csv")
wget("https://raw.githubusercontent.com/junghanss/curso_itba/master/TEO.csv")
wget("https://raw.githubusercontent.com/junghanss/curso_itba/master/TS.csv")
wget("https://raw.githubusercontent.com/junghanss/curso_itba/master/MELI.csv")
```

```python
# Parte 1 (Seleccion de empresas + Graficos historicos normales comparativos)
companies=["EDN (ADR Edenor)", "YPF (ADR YPF)", "BBAR (ADR Banco Frances)", "BMA (ADR Banco Macro)", "GGAL (ADR Galicia)", "IRS (ADR Irsa)", "CRESY (ADR Cresud)", "TEO (ADR Telecom)", "TS (ADR Tenaris)", "MELI (ADR MercadoLibre)"]         # Lista que informa al usuario sobre los nombres (tickers) de las empresas disponibles para analizar
print("Programa de análisis financiero integral de empresas" + "\n" + "\n" +"El listado de empresas disponibles para analizar son las siguientes:" + "\n", ", ".join(companies))
print("El período de información comprendida es entre 04/01/2010 al 29/04/2020" + "\n""\n" + "Elegí tres empresas, con el ticker que figura en el listado y con formato '.csv' a la hora de ingresarla en pantalla. Ejemplo: 'YPF.csv'")    # Aqui el usuario deberia poder ingresar un simple nombre para elegir la empresa (sin mayores detalles) pero no llegue a codear algo así, disculpas. Debe ser el "ticker" + ".csv"

ai=[]                   # Lista donde almacenaremos las empresas elegidas
i=1
while i <= 3:           # Armamos un loop muy simple para el ingreso de las empresas elegidas. Aqui faltaria codigo para cubrir la posibilidad que el usuario ingrese un nombre sin el ".csv" o un dato no string (no llegue con el tiempo para codearlo, disculpas)
    b= input("Elegí una empresa: ")
    ai.append(b)
    i=i+1
print("Las compañías seleccionadas son: ", ai, "\n")  # Le imprimimos en pantalla al usuario las empresas elegidas
a1=pd.read_csv(ai[0], parse_dates=["Date"], index_col="Date") # Definimos las variables a usar, que contendran el archivo (indexado por fechas) leido de la empresa que corresponda segun el indice de la lista
a2=pd.read_csv(ai[1], parse_dates=["Date"], index_col="Date")
a3=pd.read_csv(ai[2], parse_dates=["Date"], index_col="Date")
```

```python
plt.style.use("seaborn")
plt.rc('figure', figsize=(12,6))
plt.plot(a1["Close"], color="#06314d", label= ai[0])   # Graficamos el comportamiento historico de los precios de las empresas seleccionadas
plt.plot(a2["Close"], color="#91162f", label= ai[1])
plt.plot(a3["Close"], 'g', label= ai[2])    
plt.gcf().autofmt_xdate()    # Llamamos a la figura (con gcf: get current figure) para auto-ajustar (autofmt) el eje X. Automaticamente las posiciones de fechas se ajustaran segun el zoom o posicion del grafico
date_format = mpl_dates.DateFormatter('%b, %d %Y') # (b:month, d:day, Y:year) seteamos una variable con el formato elegido
plt.gca().xaxis.set_major_formatter(date_format)   # Llamamos al eje X (gca: get current axis) y le asignamos como formato a la variable anterior
plt.title("Cotizacion en periodo 01-2010 - 05-2020", fontsize=16)
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.legend()
plt.show()
```

```python
# Parte 2 (Derivadas discretas y analisis de crecimiento):
def derivada(x):
    return [x[i]-x[i-1] for i in range(1,len(x))]  # Definimos nuestra funcion para realizar la derivada discreta en el rango correspondiente

h1 = derivada(a1["Close"])
h1.append(0) 
h2 = derivada(a2["Close"])
h2.append(0)
h3 = derivada(a3["Close"])
h3.append(0)             # Aplicamos la funcion sobre los datos que deseamos y al resultado, le agregamos un simple int(0) para igualar la cantidad de datos del eje X (fechas: n) y del eje Y (diferencias: n-1)
plt.subplot(2,2,1)       # Los precios sin tendencia obtenidos a partir de la funcion de derivada los graficaremos en una sola figura, con subplots.
plt.plot_date(a1.index,h1, linestyle='solid', markersize=0.5)
plt.xlabel("Tiempo")
plt.ylabel("Variacion")
plt.title(label=ai[0])                      # Nombramos, para cada caso, los ejes y con un título el Ticker correspondiente a cada empresa para identificarla
plt.subplot(2,2,2)
plt.plot_date(a2.index,h2, linestyle='solid', markersize=0.5)
plt.xlabel("Tiempo")
plt.ylabel("Variacion")
plt.title(label=ai[1])
plt.subplot(2,2,3)
plt.gcf().tight_layout(pad=3.0)             # Se le asigna separación entre subplots para evitar una superposición entre la etiqueta del X axis con el título del inferior
plt.plot_date(a3.index,h3, linestyle='solid', markersize=0.5)
plt.suptitle("Variacion historica con derivadas discretas", fontsize=16)
plt.xlabel("Tiempo")
plt.ylabel("Variacion")
plt.title(label=ai[2])
plt.show()
```

```python
archivo_a1 = pd.read_csv(ai[0])  # Para el analisis cuantitativo de los variacoines, definimos otras variables a usar, que contendran el archivo leido de la empresa que corresponda segun el indice de la lista
archivo_a2 = pd.read_csv(ai[1])
archivo_a3 = pd.read_csv(ai[2])
a1_dict = archivo_a1.to_dict("records")     # Se crean los diccionarios con los archivos leídos sin indexar, en crudo (a diferencia de los primeros archivos con parse_dates e indexados por fechas)
a2_dict = archivo_a2.to_dict("records")
a3_dict = archivo_a3.to_dict("records")
close_a1 = []   # Definimos las listas (vacias) para los precios de cierre historicos de las empresas
close_a2 = []
close_a3 = []
```

```python
for x in a1_dict:
    close_a1.append(x["Close"])     # Definimos las iteraciones para agregar en las listas de precios de cierre, aquellos valores historicos correspondientes
for x in a2_dict:
    close_a2.append(x["Close"])
for x in a3_dict:
    close_a3.append(x["Close"])

inc_a1 = derivada(close_a1)  # Definimos las derivadas discretas de los precios dentro de una variable que representará al crecimiento de la empresa
inc_a2 = derivada(close_a2)
inc_a3 = derivada(close_a3)
start_index_ai = -1     # Definimos variables para usar como comienzo/finalizacion de los indices
end_index_ai = -1
start_index2_ai = -1
end_index2_ai = -1
start_index3_ai = -1
end_index3_ai = -1

for i in range(len(a1_dict)):                       # Definimos las iteraciones para matchear las fechas (de la consigna) de las derivadas discretas
    if a1_dict[i]["Date"] == "2020-03-01":          
        start_index_ai = i
    if a1_dict[i]["Date"] == "2020-04-01":          # Período: marzo 2020
        end_index_ai = i
    if a1_dict[i]["Date"] == "2020-02-03":          
        start_index2_ai = i
    if a1_dict[i]["Date"] == "2020-02-28":          # Período: febrero 2020
        end_index2_ai = i
    if a1_dict[i]["Date"] == "2019-04-30":          
        start_index3_ai = i
    if a1_dict[i]["Date"] == "2020-04-29":          # Período: mayo 2019 - mayo 2020 (anual)
        end_index3_ai = i
```

```python
suma_a1 = 0
suma_a2 = 0
suma_a3 = 0

for i in range(start_index_ai, end_index_ai):   # Periodo marzo # Armamos un loop para que en el rango de fechas correspondiente, el numero "suma_ai" pueda incorporar los datos de la variable crecimiento (para ese rango de fechas)
    suma_a1 = suma_a1 + inc_a1[i]
    suma_a2 = suma_a2 + inc_a2[i]
    suma_a3 = suma_a3 + inc_a3[i]
top_crec=max(suma_a1, suma_a2, suma_a3)         # La variable top_crec refiere al mayor de los valores de todas las empresas, para determinar cual fue la que más creció en el período correspondiente
if top_crec == suma_a1:
    print(f"La empresa con mayor crecimiento para el período MARZO fue {ai[0]}")  # Una vez definida cuál creció más, imprimimos en pantalla esa información con una serie de condicionales muy simples
elif top_crec == suma_a2:
    print(f"La empresa con mayor crecimiento para el período MARZO fue {ai[1]}")
else:
    print(f"La empresa con mayor crecimiento para el período MARZO fue {ai[2]}")
bottom_crec = min(suma_a1, suma_a2, suma_a3)    # Repetimos el proceso anterior para ver cuál creció menos (con la variable bottom_crec)
if bottom_crec == suma_a1:
    print(f"La empresa con menor crecimiento para el período MARZO fue {ai[0]}", "\n")
elif bottom_crec == suma_a2:
    print(f"La empresa con menor crecimiento para el período MARZO fue {ai[1]}", "\n")
else:
    print(f"La empresa con menor crecimiento para el período MARZO fue {ai[2]}", "\n")
marzo = pd.DataFrame([['Marzo:',top_crec, bottom_crec]],
                   columns=['Periodo acumulado', 'Mayor crecimiento', 'Menor crecimiento']) # Por medio de este método de pandas, definimos un DataFrame que contendrá 1 fila con los datos de marzo-valor-valor y otra fila correspondiente al indice (columns)
# print("A1: ", suma_a1, "\n","A2: ", suma_a2, "\n","A3 :", suma_a3)  # Le podriamos imprimir en pantalla al usuario la información de "suma" de cada una, a modo de verificacion.
```

```python
suma_a1 = 0     # Redefinimos el número suma otra vez en cero y comenzamos el mismo proceso llevado a cabo anteriormente.
suma_a2 = 0
suma_a3 = 0

for i in range(start_index2_ai, end_index2_ai):   #Periodo febrero
    suma_a1 = suma_a1 + inc_a1[i]
    suma_a2 = suma_a2 + inc_a2[i]
    suma_a3 = suma_a3 + inc_a3[i]
top_crec=max(suma_a1, suma_a2, suma_a3)
if top_crec == suma_a1:
    print(f"La empresa con mayor crecimiento para el período FEBRERO fue {ai[0]}")
elif top_crec == suma_a2:
    print(f"La empresa con mayor crecimiento para el período FEBRERO fue {ai[1]}")
else:
    print(f"La empresa con mayor crecimiento para el período FEBRERO fue {ai[2]}")
bottom_crec = min(suma_a1, suma_a2, suma_a3)
if bottom_crec == suma_a1:
    print(f"La empresa con menor crecimiento para el período FEBRERO fue {ai[0]}", "\n")
elif bottom_crec == suma_a2:
    print(f"La empresa con menor crecimiento para el período FEBRERO fue {ai[1]}", "\n")
else:
    print(f"La empresa con menor crecimiento para el período FEBRERO fue {ai[2]}", "\n")
febrero = pd.DataFrame([['Febrero:',top_crec, bottom_crec]],
                   columns=['Periodo acumulado', 'Mayor crecimiento', 'Menor crecimiento']) # Volvemos a definir un DataFrame para la informacion de febrero, con las mismas caracteristicas (filas y columnas)
# print("A1: ", suma_a1, "\n","A2: ", suma_a2, "\n","A3 :", suma_a3)  # Le podriamos imprimir en pantalla al usuario la información de "suma" de cada una, a modo de verificacion.
```

```python
suma_a1 = 0
suma_a2 = 0
suma_a3 = 0

for i in range(start_index3_ai, end_index3_ai):   # Periodo 2019-2020 (ultimos 12 meses)
    suma_a1 = suma_a1 + inc_a1[i]
    suma_a2 = suma_a2 + inc_a2[i]
    suma_a3 = suma_a3 + inc_a3[i]
top_crec=max(suma_a1, suma_a2, suma_a3)
if top_crec == suma_a1:
    print(f"La empresa con mayor crecimiento en los últimos 12 meses fue {ai[0]}")
elif top_crec == suma_a2:
    print(f"La empresa con mayor crecimiento en los últimos 12 meses fue {ai[1]}")
else:
    print(f"La empresa con mayor crecimiento en los últimos 12 meses fue {ai[2]}")
bottom_crec = min(suma_a1, suma_a2, suma_a3)
if bottom_crec == suma_a1:
    print(f"La empresa con menor crecimiento en los últimos 12 meses fue {ai[0]}", "\n")
elif bottom_crec == suma_a2:
    print(f"La empresa con menor crecimiento en los últimos 12 meses fue {ai[1]}", "\n")
else:
    print(f"La empresa con menor crecimiento en los últimos 12 meses fue {ai[2]}", "\n")
year = pd.DataFrame([['Anual:',top_crec, bottom_crec]],
                   columns=['Periodo acumulado', 'Mayor crecimiento', 'Menor crecimiento']) # Definimos el último DataFrame con las mismas características (filas y columnas)
# print("A1: ", suma_a1, "\n","A2: ", suma_a2, "\n","A3 :", suma_a3)  # Le podriamos imprimir en pantalla al usuario la información de "suma" de cada una, a modo de verificacion.
```

```python
workbook=pd.concat([marzo, febrero, year])  # Con una variable nueva y por medio del metodo concatenar, podemos unir los valores de los dataframes respetando el indice (columns) en un dataframe nuevo
workbook.to_csv("Resultados.csv")   # La información de este DataFrame anterior será almacenada y exportada en un excel con formato .csv bajo el nombre "Resultados"
print(workbook) # Le imprimimos al usuario la información del DataFrame final con el resumen correspondiente
```
